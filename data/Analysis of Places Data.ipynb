```python
ones = pd.read_csv('./Quarter/nyc.csv')
zeros = pd.read_csv('./Quarter/zed.csv')

df = pd.concat([ones, zeros])
df
```

```python
from patsy import dmatrices
import statsmodels.api as sm
```

```python
y, X = dmatrices('expected ~ day_name + month_name + year + hour + tempf + vis + num_establishments + num_liquor_stores + num_restaurants + num_stores + num_bars', data=df, return_type='dataframe')
```

```python
mod = sm.OLS(y, X)
res = mod.fit()
print res.summary()
```

```python
y, X = dmatrices('expected ~ num_establishments + num_liquor_stores + num_restaurants + num_stores + num_bars', data=df, return_type='dataframe')
mod = sm.OLS(y, X)
res = mod.fit()
print res.summary()

```
