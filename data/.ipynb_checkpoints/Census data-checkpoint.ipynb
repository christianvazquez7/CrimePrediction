```python
ones = pd.read_csv('./Quarter/nyc.csv')
zeroes = pd.read_csv('./Quarter/zed.csv')
```

```python
df = pd.concat([ones, zeroes])
```

```python
df10 = df.drop(df[df.year != 2010].index)
df10.max()
```

```python
ones10 = ones.drop(ones[ones.year != 2010].index)
```

```python
ones10.max()
```

```python
#collapse by geoID and year, keep expected
geoid_dict = {}
for index, row in ones10.iterrows():
    if row.geoId not in geoid_dict:
        print row.geoId, 'not in dict'
        geoid_dict[row.geoId] = { 'geoId': row.geoId, 'expected': row.expected }
    else:
        val = geoid_dict[row.geoId]['expected']
        geoid_dict[row.geoId].expected = val + row.expected
        
geoid_dict
```
