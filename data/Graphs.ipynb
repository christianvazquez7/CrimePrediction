```python
grids = pd.read_json('../src/Preprocessing/PlacesData/grids.json')
grids_list = grids[0].tolist()
```

```python
q_score = pd.read_csv('./StreetScoreNYC-Aggregated.csv')
```

```python
q_score_sorted = q_score[q_score['gridID'].isin(grids_list)]
q_score_sorted
```

```python
robbery = pd.read_csv('./RobberyNYC-Mapped.csv')
robbery
```

```python
robbery_2014 = robbery.drop(robbery[robbery.year != 2014.0].index)
robbery_2014
```

```python
#number of crimes per grid in 2014, compare with q_score for each grid

#drop columns
robbery_2014.drop(robbery_2014.columns[[1,2,3,4,5,6,7,9,10]], axis=1, inplace=True)

#aggregate
robbery_2014 = robbery_2014.groupby(robbery.gridID).sum()
robbery_2014
```

```python
#rename grid column
grids.rename(columns={0:'gridID'}, inplace=True)
grids['expected'] = 0

list(grids.columns.values)
```

```python
robbery_2014 = pd.merge(robbery_2014, grids, on='gridID', how='outer')
robbery_2014
```

```python
#set NaN to 0
robbery_2014.fillna(value=0, inplace=True)

#add expected columns
robbery_2014['expected'] = robbery_2014['expected_x'] + robbery_2014['expected_y']
robbery_2014
```

```python
#drop columns
robbery_2014.drop(robbery_2014.columns[[1,2]], axis=1, inplace=True)
```

```python
#add q_score to robbery_2014

robbery_q_2014 = pd.merge(left=robbery_2014, right=q_score_sorted, on='gridID')
robbery_q_2014
```

```python
from ggplot import *
```

```python
ss = ggplot(robbery_q_2014, aes('expected', 'avg')) + \
    geom_point(colour='steelblue') + \
    xlab('Expected Robberies') + ylab('Q-Score') + \
    ggtitle('Street Score vs. NYC Robberies in 2014')
    
ss.save('./graphs/street_score.png')
```
