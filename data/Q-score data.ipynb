```python
import scipy
import matplotlib as mpl
import pandas as pd

zeros = pd.read_csv('./Quarter/zed.csv')
ones = pd.read_csv('./Quarter/nyc.csv')
q = pd.read_csv('StreetScoreNYC-Aggregated.csv')

q
```

```python
df14 = ones.drop(ones[ones.year != 2014].index)

df14.max()
```

```python
ones
```

```python
d14 = {}

for i, r in df14.iterrows():
    if r['gridID'] not in d14:
        d14[r['gridID']] = { 'gridID': r['gridID'], 'expected': r['expected'] }
    else:
        val = d14[r['gridID']]['expected']
        d14[r['gridID']]['expected'] = val + r['expected']
        
d14
```

```python
df = pd.DataFrame.from_dict(d14, orient='index')
df
```

```python
df.max()
```

```python
grids = pd.read_json('../src/Preprocessing/PlacesData/grids.json')
grids_list = grids[0].tolist()
qs = q[q['gridID'].isin(grids_list)]
qs
```

```python
dfqs = df.merge(qs, how='right', on='gridID')
```

```python
dfqs
```

```python
dfqs.fillna(value='0', inplace=True)
dfqs
```

```python
from ggplot import *
```

```python
ss = ggplot(dfqs, aes('expected', 'avg')) + \
    geom_point() + \
    theme(axis_text=element_text(size=12, family='serif')) + \
    xlab(element_text(text='Robberies in 2014', size=14, family='serif')) + \
    ylab(element_text(text='Q-score', size=14, family='serif'))
```
